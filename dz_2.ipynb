**Регрессионный анализ** - можно ли провести линейную зависимость между двумя выборками. Возьмем Ladder - удовлетворение жизнью, и generosity - доброта. При проецировании данных на эту прямую разброс данных максимален?

```python
ladder = data['Ladder']
generosity = data['Generosity']
sns.regplot(x= ladder, y=generosity ,data=data, fit_reg=True)
plt.xlabel('Удовлетворение жизнью')
plt.ylabel('Доброта')
plt.title('Корреляционная зависимость двух выборок')
plt.grid(False)
plt.show()

# Насчет собственного вектора поняла не очень, но думаю, что данная прямая им не является
# Следовательно, разброс данных на прямую не является максимальным, и согласно линейной регрессии, 
# сумма расстояний от каждого элемента выборки до прямой минимальна
```

# Дискретное распределение случайной величины

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

```python
n_experiments = 10

rng = np.random.RandomState(123)
outcomes = np.empty(n_experiments, dtype=np.float)

for i in range(n_experiments):
    outcomes[i] = rng.randint(1, 7)

print('Результат моделирования подбрасываний: ', outcomes)
plt.hist(outcomes)
plt.xlabel('Грань кубика')
plt.ylabel('Количество выпадаемости')
plt.show()


# т.к. выборка небольшая, тяжело сказать, какое это распределение
# скорее всего стремится к равномерному
```

Напишите код для моделирования 1000 подбрасываний игрального кубика. Начертите диаграму histogram. Почему график изменился? Как применить центральную предельную теорему?

```python
n_experiments = 1000

rng = np.random.RandomState(123)
outcomes = np.empty(n_experiments, dtype=np.float)

for i in range(n_experiments):
    outcomes[i] = rng.randint(1, 7)

print('Результат моделирования подбрасываний: ', outcomes)
plt.hist(outcomes)
plt.xlabel('Грань кубика')
plt.ylabel('Количество выпадаемости')
plt.show()

# график изменился, т.к. стало больше данных в выборке
# соответствует равномерному распределению
```

```python
# применение центральной предельной теоремы
import scipy.special as sps
import random

# Распределение средних значений выборки размера 30 из исходного распределения является нормальным
samples = [ np.mean(random.choices(outcomes, k=30)) for _ in range(1000) ]

plt.hist(samples)
plt.title('Средние значения выборок размера 30 из исходного распределения')
plt.show()
```
