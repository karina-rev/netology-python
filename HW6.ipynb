**Задание 1**

Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

Результат классификации запишите в отдельный столбец region


```python
import pandas as pd
```

```python
keywords = pd.read_csv('keywords.csv')
keywords.head()
```

```python
def geo_class(row):
    """ Вовзращает название региона, если поисковой запрос в столбце 'keyword' содержит название города региона
    Если поисковый запрос не содержит названия города, то возвращает 'undefined'
    Правила распределения по регионам Центр, Северо-Запад и Дальний Восток хранится в словаре geo_data"""
    
    geo_data = {
        'Центр': ['москва', 'тула', 'ярославль'],
        'Северо-Запад': ['петербург', 'псков', 'мурманск'],
        'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }
    
    for region, cities in geo_data.items():
        for city in cities:
            if city in row['keyword']:
                return region
    return 'undefined'
```

```python
keywords['region'] = keywords.apply(geo_class, axis=1)
keywords.head()
```

```python
keywords[keywords['region'] != 'undefined']
```

**Задание 2**

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
    - оценка 2 и меньше - низкий рейтинг
    - оценка 4 и меньше - средний рейтинг
    - оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
```

```python
# JOIN ratings с movies
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
def rating_class(row):
    """ По значение из столбца 'rating' возвращает одно из значений:
        - оценка 2 и меньше - низкий рейтинг
        - оценка 4 и меньше - средний рейтинг
        - оценка 4.5 и 5 - высокий рейтинг """
    
    if row['rating'] <= 2:
        return 'низкий рейтинг'
    elif row['rating'] <= 4:
        return 'средний рейтинг'
    elif row['rating'] >= 4.5:
        return 'высокий рейтинг'
```

```python
joined['class'] = joined.apply(rating_class, axis=1)
joined
```

**Задание 3**

Посчитайте среднее значение Lifetime киноманов (пользователи, которые поставили 100 и более рейтингов). Под Lifetime понимается разница между максимальным и минимальным значением timestamp для каждого пользователя. Ответ дайте в днях.

```python
# количество записей по каждому юзеру
ratings.groupby('userId').count().head()
```

```python
# записываем количество записей по каждому индексу в grouped и обновляем индекс
grouped = ratings.groupby('userId').count().reset_index()
grouped.head()
```

```python
# фильтруем, оставляем только тех юзеров, которые ставили оценки >= 100
ratings_filtered = grouped[grouped['timestamp'] >= 100]
ratings_filtered.head()
```

```python
# записываем в film_fans_user_ids список отфильтрованных пользователей
film_fans_user_ids = ratings_filtered['userId'].tolist()
film_fans_user_ids
```

```python
# фильтруем ratings, оставляем только тех пользователей, которые есть в списке film_fans_user_ids
# и для кождого пользователя считаем минимальный и максимальный timestamp
users_timestamp = (ratings[ratings['userId'].isin(film_fans_user_ids)]
                   .groupby('userId')
                   .agg({'timestamp': ['min', 'max']})
                  )
users_timestamp.head()
```

```python
# добавляем столбец lifetime, куда записываем разницу между максимальным и минимальным значением timestamp
users_timestamp['lifetime'] = users_timestamp.apply(lambda x: x['timestamp']['max'] - x['timestamp']['min'], axis=1)
users_timestamp.head()
```

```python
# вычисляем среднее значение lifetimes для всех юзеров
lifetimes = users_timestamp['lifetime'].tolist()
meanLife = round(sum(lifetimes) / float(len(lifetimes)))
meanLife
```

```python
# переводим timestamp в дни
from datetime import datetime
meanLifeInDays = (datetime.utcfromtimestamp(meanLife) - datetime.fromtimestamp(0)).days
meanLifeInDays
```

**Задание 4**

Есть мнение, что "раньше снимали настоящее кино, не то что сейчас". Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из материалов занятия. Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010.

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
    - для каждой строки пройдите по всем годам списка years
    - если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
    - если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец 'year'

4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца 'year' и отсортируйте результат по убыванию рейтинга

```python
# список всех годов с 1950 по 2010
years = []
for year in range(1950, 2011, 1):
    years.append(year)
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')

# JOIN ratings с movies
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
def production_year(row):
    """ Eсли номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
    Eсли ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год"""
    
    for year in years:
        if str(year) in row['title']:
            return year
    return 1900
```

```python
joined['year'] = joined.apply(production_year, axis=1)
joined
```

```python
# Cредний рейтинг всех фильмов для каждого значения столбца 'year', отсортирован по убыванию рейтинга
(joined
 .groupby('year')
 .agg({'rating': 'mean'})
 .sort_values('rating', ascending=False)
 .head())
```
